```python
from xbbg import blp
# Pull S&P 500 constituents
tickers = blp.bds('SPX Index', 'INDX_MEMBERS')
print(tickers.head())
```

```python
# Pull EOD prices for Apple and Microsoft
prices = blp.bdh(
    tickers=['AAPL US Equity', 'MSFT US Equity'],
    flds=['PX_LAST', 'PX_VOLUME'],
    start_date='2023-01-01',
    end_date='2023-12-31'
)

print(prices)
```

```python
# Pull Annual Net Income and EPS
fundamentals = blp.bdh(
    tickers='AAPL US Equity',
    start_date='2020-01-01',
    end_date='2023-12-31',
    Per='Y' # 'Y' for Yearly, 'Q' for Quarterly
)

print(fundamentals)
```
